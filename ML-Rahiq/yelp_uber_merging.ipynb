```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
df_uber = pd.read_csv("Restaurant_File_with_address.csv")
df_uber
```

```python
df_uber_price = pd.read_csv("restaurants_price.csv")
df_uber_price
```

```python
inner_merge = pd.merge(df_uber, df_uber_price, on='Restaurant_Name', how='inner')
inner_merge
```

```python
Resturant_price_uber_df = inner_merge[["Restaurant_Name","min_price"]]
Resturant_price_uber_df.head(10)
```

```python
df_yelp = pd.read_csv("Restaurants_Yelp_With_Lat.csv")
df_yelp
```

```python
final_df = pd.merge(df_yelp, Resturant_price_uber_df, on=["Restaurant_Name","min_price"], how='left')
final_df
```

```python
final_df.to_csv('uber_yelp_merged.csv', index=False)
```

```python
final_df['min_price'] = final_df['min_price'].replace({'$': 10 , '$$': 25 ,'$$$' : 45 , '$$$$' : 50})

final_df.head(10)
```

```python
final_df.to_csv('uber_yelp_merged_with_price.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (20,20))
sns.heatmap(final_df.corr(), annot = True)
plt.show()
```
